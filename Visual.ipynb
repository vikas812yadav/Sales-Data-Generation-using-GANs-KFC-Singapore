```julia
using Pkg
using DataFrames
using CSV
using Plots
using Word2Vec
```

# Unique Transactions in each Segment

```julia
cluster = [2,1,6,4,3,3]
segment = ["Breakfast_Weekday","Breakfast_Weekend","Lunch_Weekday","Lunch_Weekend","Dinner_Weekday","Dinner_Weekend"]
for i in 1:6
    filename = segment[i] * "_Clusters.csv"
    df = DataFrame(CSV.File(filename))
    df = groupby(df,:cluster)
    for j in 1:length(df)
        s = segment[i] * "_Cluster:" * string(Int64(df[j].cluster[1])) * " - " * string(size(df[j],1))
        display(s)
    end
end
```

# Item Frequency

```julia
cluster = [2,1,6,4,3,3]
segment = ["Breakfast_Weekday","Breakfast_Weekend","Lunch_Weekday","Lunch_Weekend","Dinner_Weekday","Dinner_Weekend"]
for i in 1:6
    for j in 1:cluster[i]
        filename = segment[i] * "_Cluster-" * string(j) * "_Word_Frequency.csv"
        df = DataFrame(CSV.File(filename))
        df1 = sort(df,:Frequency,rev = true)
        #display top 20 frequent items
        df2 = df1[1:20,:]
        s = segment[i] * " Cluster:" * string(j)
        display(s)
        display(df2)
    end
end
```

# Sales Amount

```julia
cluster = [2,1,6,4,3,3]
segment = ["Breakfast_Weekday","Breakfast_Weekend","Lunch_Weekday","Lunch_Weekend","Dinner_Weekday","Dinner_Weekend"]
for i in 1:6
    for j in 1:cluster[i]
        path = "Bundle/" * segment[i] * "_Clusters_sales" * string(j) * ".csv"
        df = DataFrame(CSV.File(path))
        df = df[((df.sales_amount .>= 0) .& (df.sales_amount .<= 50)),:]
        display(histogram(df.sales_amount,bins=100,title = (segment[i] * " Cluster:" * string(j) * " wrt Sales")))
    end
end
```

# Sales Channel

```julia
cluster = [2,1,6,4,3,3]
segment = ["Breakfast_Weekday","Breakfast_Weekend","Lunch_Weekday","Lunch_Weekend","Dinner_Weekday","Dinner_Weekend"]
for i in 1:6
    path = "Bundle/" * segment[i] * "_channel_vs_cluster.csv"
    df = DataFrame(CSV.File(path))
    for j in 1:cluster[i]
        display(
        bar(df.i,
        df[:,j+1],
        label = "Frequency/Sales_Channel",
        title = segment[i] * " Cluster:" * string(j) * " wrt Sales_Channel",
        xticks =:all,
        xrotation = 45,
        legend =:topright))
    end
end
```
